```python
# import library 
import numpy as np
import skfuzzy as fuzz
from matplotlib import pyplot as plt
```

```python
#  generate variable global
x_production = np.arange(0,10.01,0.5)
x_smartphone = np.arange(0,10.01,0.5)
x_tip = np.arange(0,25.01,0.5)
```

```python
# generate membership function untuk ketiga variable yang ada 
# pada rentang tinggi, sedang dan rendah
production_low = fuzz.trimf(x_production, [0, 3, 5])
production_middle = fuzz.trimf(x_production, [0,8, 10])
production_high = fuzz.trimf(x_production, [7, 10, 10])
# fungsi keanggotaan servis
smartphone_low = fuzz.zmf(x_smartphone, 0, 5)
smartphone_middle = fuzz.pimf(x_smartphone, 0, 5, 5, 10)
smartphone_high = fuzz.smf(x_smartphone, 5, 10)
# jumlah nilai servis dan smartphone
production_score = 9.5
smartphone_score = 5.0
```

```python
# memberikan nilai 4.0 untuk kualitas makanan dan 9.5 untuk kualitas pelayanan
production_score = 9.5
smartphone_score = 4.0

production_low_degree = fuzz.interp_membership(
    x_production, production_low, production_score)
production_middle_degree = fuzz.interp_membership(
    x_production, production_middle, production_score)
production_high_degree = fuzz.interp_membership(
    x_production, production_high, production_score)

smartphone_low_degree = fuzz.interp_membership(x_smartphone, smartphone_low, smartphone_score)
smartphone_middle_degree = fuzz.interp_membership(x_smartphone, smartphone_middle, smartphone_score)
smartphone_high_degree = fuzz.interp_membership(x_smartphone, smartphone_high, smartphone_score)
```

```python
# Whole config
fig_scale_x =2.0
fig_scale_y =1.5
fig = plt.figure(figsize=(6.4 * fig_scale_x,4.8 * fig_scale_y))
row = 2
col = 3

plt.subplot(row, col, 1)
plt.title("Production Quality")
plt.plot(x_production, production_low, label="low", marker=".")
plt.plot(x_production, production_middle, label="middle", marker=".")
plt.plot(x_production, production_high, label="high", marker=".")
plt.plot(production_score, 0.0, label="production score", marker="D")
plt.plot(production_score,production_low_degree,
    label="low degree", marker="o")
plt.plot(production_score,production_middle_degree,
    label="middle degree", marker="o")
plt.plot(production_score,production_high_degree,
    label="high degree", marker="o")
plt.legend(loc="upper left")

plt.subplot(row, col, 2)
plt.title("Smartphone Quality")
plt.plot(x_smartphone, smartphone_low, label="low", marker=".")
plt.plot(x_smartphone, smartphone_middle, label="middle", marker=".")
plt.plot(x_smartphone, smartphone_high, label="high", marker=".")
plt.plot(smartphone_score,0.0, label="smartphone score", marker="D")
plt.plot(smartphone_score, smartphone_low_degree, label="low degree", marker="o")
plt.plot(smartphone_score, smartphone_middle_degree, label="middle degree", marker="o")
plt.plot(smartphone_score, smartphone_high_degree, label="high degree", marker="o")
plt.legend(loc="upper left")
```
